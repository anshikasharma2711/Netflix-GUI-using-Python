# Content Type on Netflix

```python
# Calculate the percentage distribution of 'type' column
type_counts = netflix['type'].value_counts(normalize=True) * 100

# Define Netflix color palette (using official brand colors or similar)
netflix_red = '#E50914'
netflix_black = '#221f1f'
netflix_colors = [netflix_red, netflix_black]

# Specify explode settings to emphasize a section (e.g., 'TV Shows')
explode = (0.1, 0)  # Explode the first slice (TV Shows)

# Plotting the pie chart with 3D effect and explode
plt.figure(figsize=(8, 6))  # Adjust the figure size
wedges, _ = plt.pie(type_counts, labels=None, startangle=140,
                    colors=netflix_colors, explode=explode, shadow=True,
                    wedgeprops={'edgecolor': 'white', 'linewidth': 2, 'antialiased': True})

plt.title('Distribution of Content Types on Netflix', fontsize=14, fontweight='bold', color=netflix_black, pad=20)

# Add custom legend labels at the side
plt.legend(wedges, type_counts.index, title='Content Type', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))

# Add labels for 'TV Shows' and 'Movies' next to the pie chart
plt.text(-1, 0.5, 'TV Shows', fontsize=12, fontweight='bold', color=netflix_black)
plt.text(-1, -0.5, 'Movies', fontsize=12, fontweight='bold', color=netflix_black)

# Display the pie chart without text labels inside the pie
plt.show()
```

# Growth in content over the years

```python
# Filter data by 'type' (Movie or TV Show) and 'release_year'
movie_data = netflix[netflix['type'] == 'Movie'].groupby('release_year').size()
tv_show_data = netflix[netflix['type'] == 'TV Show'].groupby('release_year').size()

# Create a stacked area plot with Netflix-like colors
plt.figure(figsize=(12, 8))

# Plot cumulative number of Movies over the years with Netflix red color (#E50914)
plt.fill_between(movie_data.index, movie_data.cumsum(), label='Movies', color='#E50914', alpha=0.7)

# Plot cumulative number of TV Shows over the years with Netflix black color (#000000)
plt.fill_between(tv_show_data.index, tv_show_data.cumsum(), label='TV Shows', color='#000000', alpha=0.7)

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Cumulative Number of Titles')
plt.title('Cumulative Growth of Movies and TV Shows Over Years on Netflix')
plt.legend()

# Show the plot
plt.show()
```

# Content added over month

```python
# Convert 'date_added' column to datetime format
netflix['date_added'] = pd.to_datetime(netflix['date_added'])

# Extract month from 'date_added' and count the number of titles added in each month
netflix['added_month'] = netflix['date_added'].dt.month
monthly_counts = netflix['added_month'].value_counts().sort_index()

# Map month numbers to month names for better readability on the plot
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_counts.index = [month_names[int(month)-1] for month in monthly_counts.index]

# Define Netflix-themed colors (red and black)
netflix_red = '#E50914'
netflix_black = '#221f1f'

# Create a bar graph with Netflix-themed colors
plt.figure(figsize=(12, 8))

# Plot the bar chart with custom colors
plt.bar(monthly_counts.index, monthly_counts.values, color=[netflix_red]*6 + [netflix_black]*6)

# Set labels and title
plt.xlabel('Month')
plt.ylabel('Number of Titles Added')
plt.title('Content added over Month')

# Show the plot
plt.show()
```

# Top 10 countries producing TV shows & Movies respectively

```python
# Remove rows with missing 'country' values
netflix = netflix.dropna(subset=['country'])

# Separate TV shows and movies based on 'type' column
tv_shows = netflix[netflix['type'] == 'TV Show']
movies = netflix[netflix['type'] == 'Movie']

# Count the occurrences of each country for TV shows
tv_show_countries = tv_shows['country'].value_counts().head(10)

# Count the occurrences of each country for movies
movie_countries = movies['country'].value_counts().head(10)

# Create subplots for side-by-side comparison
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Plot for top 10 countries producing TV shows
axes[0].barh(tv_show_countries.index, tv_show_countries.values, color='#E50914')
axes[0].set_title('Top 10 Countries Producing TV Shows')
axes[0].set_xlabel('Number of TV Shows')
axes[0].invert_yaxis()  # Invert y-axis for better visualization

# Plot for top 10 countries producing movies
axes[1].barh(movie_countries.index, movie_countries.values, color='black')
axes[1].set_title('Top 10 Countries Producing Movies')
axes[1].set_xlabel('Number of Movies')
axes[1].invert_yaxis()  # Invert y-axis for better visualization

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

# Actor appearance in movies by countries

```python
# Filter the dataset to include only movies
movies = netflix[netflix['type'] == 'Movie']

# Count the number of movies produced by each country
movie_countries_counts = movies['country'].value_counts()

# Select the top 5 countries producing the most movies
top_countries = movie_countries_counts.head(5).index.tolist()

# Initialize a dictionary to store top actors for each country
top_actors_by_country = {}

# Iterate over each top country and find the top actors with the most movie appearances
for country in top_countries:
    country_movies = movies[movies['country'] == country]
    actors_counts = country_movies['cast'].str.split(', ', expand=True).stack().value_counts()
    top_actors = actors_counts.head(10)  # Select top 10 actors
    top_actors_by_country[country] = top_actors

# Combine top actors across countries into a single DataFrame
combined_top_actors = pd.concat(top_actors_by_country.values(), axis=1)
combined_top_actors.columns = top_actors_by_country.keys()

# Create a heatmap of actor appearances across countries
plt.figure(figsize=(12, 8))
sns.heatmap(combined_top_actors.fillna(0), cmap='YlGnBu', annot=False, linewidths=.5)
plt.title('Actor Appearances in Movies by Country')
plt.xlabel('Country')
plt.ylabel('Actor')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Actor appearance in TV shows by countries

```python
# Filter the dataset to include only TV shows
tv_shows = netflix[netflix['type'] == 'TV Show']

# Count the number of TV shows produced by each country
tv_countries_counts = tv_shows['country'].value_counts()

# Select the top 5 countries producing the most TV shows
top_countries = tv_countries_counts.head(5).index.tolist()

# Initialize a dictionary to store top actors for each country in TV shows
top_actors_by_country_tv = {}

# Iterate over each top country and find the top actors with the most TV show appearances
for country in top_countries:
    country_tv_shows = tv_shows[tv_shows['country'] == country]
    actors_counts_tv = country_tv_shows['cast'].str.split(', ', expand=True).stack().value_counts()
    top_actors_tv = actors_counts_tv.head(10)  # Select top 10 actors
    top_actors_by_country_tv[country] = top_actors_tv

# Combine top actors across countries into a single DataFrame for TV shows
combined_top_actors_tv = pd.concat(top_actors_by_country_tv.values(), axis=1)
combined_top_actors_tv.columns = top_actors_by_country_tv.keys()

# Create a heatmap of actor appearances across countries for TV shows
plt.figure(figsize=(12, 8))
sns.heatmap(combined_top_actors_tv.fillna(0), cmap='YlGnBu', annot=False, linewidths=.5)
plt.title('Actor Appearances in TV Shows by Country')
plt.xlabel('Country')
plt.ylabel('Actor')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
